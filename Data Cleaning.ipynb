```python
# Replace Outliers with upper & Lower fence

# Function to outliers
def Handle_Outliers(real, column):
    Q1 = real[column].quantile(0.25)
    Q3 = real[column].quantile(0.75)
    IQR = Q3 - Q1
    Lower_fence = Q1 - 1.5 * IQR
    Upper_fence = Q3 + 1.5 * IQR
    real.loc[real[column] < Lower_fence, column] = Lower_fence
    real.loc[real[column] > Upper_fence, column] = Upper_fence

# Select numeric columns
numeric_real = real.select_dtypes(include=['number'])

# Apply to all numeric columns
for column in numeric_real:
    Handle_Outliers(real, column)
```

```python
numeric_real = real.select_dtypes(include=['number'])

for col in numeric_real:
    Q1 = real[col].quantile(0.25)
    Q3 = real[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_fence = Q1 - 1.5 * IQR
    upper_fence = Q3 + 1.5 * IQR

    outliers_count = ((real[col] < lower_fence) | (real[col] > upper_fence)).sum()

    if outliers_count == 0:
        print(f"{col}: No outliers")
    else:
        print(f"{col}: {outliers_count} outliers remaining")
```
